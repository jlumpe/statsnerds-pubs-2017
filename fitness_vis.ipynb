```python
from __future__ import division
import cPickle as pic
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
fitness1 = pic.load(open("fitness_ym_1.2.pkl", "rb"))
fitness2 = pic.load(open("fitness_ym_2.2.pkl", "rb"))
```

```python
fit_vals1 = []
fit_vals2 = []
for bc in fitness1:
    if fitness2.has_key(bc):
        fit_vals1.append(fitness1[bc][0])
        fit_vals2.append(fitness2[bc][0])


fit_vals1 = np.array(fit_vals1)
fit_vals2 = np.array(fit_vals2)
```

```python
x_vals = np.arange(10, max(fit_vals1), step=.01)
y_vals = x_vals


spcorr = stats.spearmanr(fit_vals1, fit_vals2)
pecorr = stats.pearsonr(fit_vals1, fit_vals2)
print(spcorr)
print(pecorr)

plt.plot(fit_vals1, fit_vals2, 'ro', alpha=.5)
plt.xlabel("Fitness Rep 1")
plt.ylabel("Fitness Rep 2")
plt.xlim(0,1)
plt.ylim(0,1)
plt.plot(x_vals, y_vals, 'b-')
plt.title("Fitness Comparison Between Replicates; Pearson: .98")
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(fit_vals1, fit_vals2)
print(slope)

fn = np.array(fit_vals1) * slope + intercept
print(np.sqrt(np.mean((fit_vals2 - fn)**2)))
```
